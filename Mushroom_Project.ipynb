```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

### Finding best random state

```python
from sklearn.linear_model import LogisticRegression
MaxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=i)
    mod=LogisticRegression()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>MaxAccu:
        MaxAccu=acc
        maxRS=i
print("Best accuracy is",MaxAccu,'on radom state',maxRS)
```

### Training and splitting the data

```python
# Training and Splitting the data  
x_train,x_test,y_train,_y_test=train_test_split(x,y,test_size=0.3,random_state=21)
print(y_train)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python

# Using Logistic regression classification model
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

```python
#Using Randome forest classifier
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
pred=RF.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

```python
from sklearn.model_selection import cross_val_score
```

### Saving the model

```python
Final_mod2=RandomForestClassifier(criterion='gini',max_depth=5,max_features='auto',n_estimators=200)
Final_mod2.fit(x_train,y_train)
pred=Final_mod2.predict(x_test)
acc=accuracy_score(y_test,pred)
print(acc*100)
```
